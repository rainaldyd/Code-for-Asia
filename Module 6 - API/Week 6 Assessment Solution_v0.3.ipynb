```python
import pandas as pd
import requests
import json
import matplotlib.pyplot as plt
```

```python
## Q1
# Make request using GET

from_date = "2022-01-01T00:01Z"
to_date = "2022-02-01T00:00Z"

url = f"https://api.carbonintensity.org.uk/intensity/{from_date}/{to_date}"
r = requests.get(url)
intensity = r.json()
display(intensity)
```

```python
## Q2
# Store data into a Dataframe

carbon_df = pd.json_normalize(intensity["data"])
display(carbon_df.head())
display(carbon_df.tail())
```

```python
## Q3
# Simple clean & wrangle dataset

# Rename
carbon_df.rename(columns={"from":"From_Date",
                          "to":"To_Date",
                          "intensity.forecast":"Carbon_Forecast",
                          "intensity.actual":"Actual_Carbon_Intensity",
                          "intensity.index":"Carbon_Index"},
                 inplace=True)
```

```python
# Convert data type
carbon_df.info()
```

```python
carbon_df = carbon_df.astype({"From_Date":"datetime64",
                              "To_Date":"datetime64"},
                             errors = "ignore")
carbon_df.dtypes
```
